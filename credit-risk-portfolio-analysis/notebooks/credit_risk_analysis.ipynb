# Credit Risk Portfolio Analysis
This notebook loads the synthetic lending dataset, performs exploratory analysis, and prepares summary outputs for dashboarding.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path.cwd().resolve().parents[0] if 'notebooks' in str(Path.cwd()) else Path.cwd()
DATA = ROOT / 'data' / 'raw' / 'loan_data.csv'
df = pd.read_csv(DATA)
df.head()
```

```python
df.shape, df.isna().sum().sort_values(ascending=False).head(10)
```

```python
df['dti_bucket'] = pd.cut(df['dti_ratio'], bins=[0,10,20,30,50], labels=['0-10','10-20','20-30','30+'])
df['util_bucket'] = pd.cut(df['utilization_rate'], bins=[0,30,50,70,100], labels=['0-30','30-50','50-70','70+'])
summary_grade = df.groupby('grade', as_index=False).agg(total_loans=('customer_id','count'), default_rate=('default_flag','mean'), avg_loan=('loan_amount','mean'))
summary_grade['default_rate'] = (summary_grade['default_rate']*100).round(2)
summary_grade.sort_values('grade')
```

```python
plt.figure(figsize=(8,4))
plt.bar(summary_grade['grade'], summary_grade['default_rate'])
plt.title('Default Rate by Grade')
plt.xlabel('Grade')
plt.ylabel('Default Rate (%)')
plt.tight_layout()
plt.show()
```

```python
summary_purpose = df.groupby('purpose', as_index=False).agg(default_rate=('default_flag','mean')).sort_values('default_rate', ascending=False)
summary_purpose['default_rate'] = (summary_purpose['default_rate']*100).round(2)
summary_purpose
```
