# Insurance Claims & Fraud Analytics
This notebook explores fraud risk patterns in the synthetic insurance claims dataset.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path.cwd().resolve().parents[0] if 'notebooks' in str(Path.cwd()) else Path.cwd()
DATA = ROOT / 'data' / 'raw' / 'insurance_claims.csv'
df = pd.read_csv(DATA, parse_dates=['claim_date'])
df.head()
```

```python
df['tenure_bucket'] = pd.cut(df['policy_tenure_years'], bins=[0,2,5,10,20], labels=['0-2','3-5','6-10','11+'])
incident_summary = df.groupby('incident_type', as_index=False).agg(claims=('claim_id','count'), fraud_rate=('fraud_reported','mean'), avg_claim=('claim_amount','mean')).sort_values('fraud_rate', ascending=False)
incident_summary['fraud_rate'] = (incident_summary['fraud_rate']*100).round(2)
incident_summary
```

```python
plt.figure(figsize=(8,4))
plt.bar(incident_summary['incident_type'], incident_summary['fraud_rate'])
plt.title('Fraud Rate by Incident Type')
plt.xlabel('Incident Type')
plt.ylabel('Fraud Rate (%)')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```
